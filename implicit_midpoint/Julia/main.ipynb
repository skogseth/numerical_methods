# Setup

```julia
using Plots
```

```julia
include("src/newtons_method.jl")
```

# Calculation

```julia
g(x) = x^2 + 4*x + 3
dg(x) = 2*x + 4

nullpoint = newtons_method(0., g, dg)
```

# Plotting

```julia
x = range(-10, 10, 1001)
y = g.(x)

plt = plot()
plt = plot!(x, y)
plt = scatter!([nullpoint], [0.])
```
